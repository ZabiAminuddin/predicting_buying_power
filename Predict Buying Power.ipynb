# Step 4: Training the Model

```python
from sklearn.preprocessing import MinMaxScaler
scalar = MinMaxScaler()
x_scaled = scalar.fit_transform(X)
y = y.values.reshape(-1,1)
```

```python
x_scaled
```

```python
y_scaled = scalar.fit_transform(y)
```

```python

y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x_scaled, y_scaled)
# explicitly define the test percentage eg: 30%
X_train, X_test, y_train, y_test = train_test_split(x_scaled, y_scaled, test_size=0.15)
```

```python
x_scaled.shape
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
import tensorflow.keras as keras
from keras.layers import Dense
from keras.models import Sequential

model = Sequential()
model.add(Dense(25, input_dim = 5, activation='relu'))
model.add(Dense(25, activation='relu'))
model.add(Dense(1, activation='linear'))
model.summary()
```

```python
model.compile(optimizer='adam', loss='mean_squared_error')
epoc_hist = model.fit(X_train, y_train, epochs = 100, batch_size = 25, verbose = 1, validation_split=0.2)
```

```python
epoc_hist.history.keys()
```

```python
plt.plot(epoc_hist.history['loss'])
plt.plot(epoc_hist.history['val_loss'])
plt.title("Model loss progress")
plt.ylabel("Trainig and Validation Loss")
plt.xlabel("Epoch Number")
```

```python
# Lets see what the system will predict
# Gender, Age, Annual_Salary, Credit Card debt, assets
X_test = np.array([[1, 50, 60000, 10000, 600000]])
y_predict = model.predict(X_test)
```
